```python
import pandas as pd
import json
stations = pd.read_json('stations.json')
```

```python
def get_stations(bounds):
    """
    Get all stations within the given bounds.

    :param bounds: The bounds of the area to search for stations.
    :type bounds: list
    :return: A list of stations.
    :rtype: list
    """
    output = []

    for station in stations.itertuples():
        print(station.id, station.lat, station.lng, sep=',')
        print(station.lng, bounds[0], station.lng, bounds[2])
        if station.lng > bounds[0] and station.lng < bounds[2]:
            if station.lat > bounds[1] and station.lat < bounds[3]:
                output.append(station)
            
    print(output)
    
    return stations

i = get_stations([-82.87167, 26.64833, -80.80833, 26.13167])
```
